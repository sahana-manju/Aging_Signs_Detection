# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
```

# CREATING FEATURES AND TARGETS

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
puffyeyes_path='E:/puf'
train_puffyeyes=load_images_from_folder(puffyeyes_path)
target=list(np.ones(len(train_puffyeyes)))
print(len(target))
print(len(train_puffyeyes))
```

```python
nopuffyeyes_path='E:/norm'
train_nopuffyeyes=load_images_from_folder(nopuffyeyes_path)
target_n=list(np.zeros(len(train_nopuffyeyes)))
print(len(train_nopuffyeyes))
print(len(target_n))
train_nopuffyeyes[5].shape
```

```python
train_features=train_puffyeyes+train_nopuffyeyes
train_targets=target+target_n
```

```python
train_features=np.array(train_features)
train_targets=np.array(train_targets)
len(train_targets)
```

# SPLIT TRAIN AND TEST SET

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(train_features,train_targets,test_size=0.2,random_state=0)
```

```python
y_train.shape
```

# CHANGING THE TARGET TO CATEGORICAL

```python
from keras.utils import to_categorical
y_train=to_categorical(y_train)
y_train.shape
```

# IMPLEMENTATION OF EFFICIENET MODEL

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras import layers
import tensorflow

inputs=layers.Input(shape=(224,224,3))
conv_base=EfficientNetB0(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
from tensorflow.keras import models
model=models.Sequential()
model.add(conv_base)
model.add(layers.GlobalMaxPooling2D(name="gap"))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.2,name="drop_out"))
model.add(layers.Dense(2,activation="softmax",name="fc_out"))
```

```python
conv_base.trainable=False
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
model.save('efficientnet.h5')
```

```python
model.summary()
```

# PREDICTING ON TEST SET

```python
y_pred=model.predict(x_test)
```

```python
pred=[]
for i in range(len(y_pred)):
    if y_pred[i][0]>=y_pred[i][1]:
        pred.append(0)
    else:
        pred.append(1)
        
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
